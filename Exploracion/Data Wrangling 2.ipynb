```python
import pandas as pd

data = pd.read_csv("properati_clean.csv")
data.sample(3)
```

```python
data['price_per_m2_decile'] = pd.cut(data.price_per_m2, bins = 10)
```

```python
from sklearn.feature_selection import chi2

categorical = ['property_type', 'place_name', 'pileta', 'galeria', 'seguridad', 'sotano', 'amenities',
               'terraza', 'metrobus', 'subte', 'cochera', 'consultorio']

subset = data.loc[:, categorical + ['price_per_m2_decile']].dropna()

chi2(subset.loc[:, categorical], subset.price_per_m2_decile)
```
