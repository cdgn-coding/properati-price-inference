## Vamos a analizar las expensas de CABA
### ¿El valor de la propiedad en CABA correlaciona con el valor de las expensas?
Para este análisis vamos a tomar aquellas propiedades que tengan el precio en dolares y el campo de las expensas no nulo

```python
data_properati['expenses'].describe()
```

```python
# voy a analizar la variable expenses de caba
expenses_not_null_caba = (data_properati['expenses'].isnull() == 0) & (data_properati['state_name'] == 'Capital Federal')
data_expenses_caba = data_properati.loc[expenses_not_null_caba, :]
data_expenses_caba
```

```python
# quiero saber como estan agrupados los datos por currency 
data_expenses_caba['currency'].value_counts()
```

```python
data_expenses_caba_dummies = pd.get_dummies(data_expenses_caba['currency'], drop_first = True, prefix = 'currency') 
data_expenses_caba_dummies
```

```python
# corroboro que currency_usd tenga la misma cantidad de valores que hay en la columna currency con el valor USD
data_expenses_caba_dummies.sum()
```

```python
# agrego la dummy al dataframe data_expenses_caba
data_expenses_caba = data_expenses_caba.join(data_expenses_caba_dummies)
```

```python
# dropeo algunas columnas del dataframe que no me interesan para el analisis de las expenses
data_expenses_caba = data_expenses_caba.drop(labels = ['geonames_id','lat-lon', 'lat', 'lon'], axis = 1)
```

```python
# quiero ver si precio en dolares y expensas correlacionan o no, en capital federal 
# genero una mascara para quedarme unicamente con dolares = 1 
data_currency_caba_usd = data_expenses_caba['currency_USD'] == 1
# filtro las columas price y expenses
data_expenses_caba_usd = data_expenses_caba.loc[data_currency_caba_usd, ['price', 'expenses']]
# aplico pd.corr para saber si correlacionan o no 
data_expenses_caba_usd.corr(method = 'pearson')
```

**Conclusión: el precio de la propiedad en CABA no correlaciona con el valor de las expensas**

## ¿El precio en dolares correlaciona con el valor de las expensas?

Para este análisis vamos a tomar aquellas propiedades que tengan el precio en dolares y el campo de las expensas no nulo, sin importar a que provincia pertencen

```python
# listo todos los state_names disponibles 
state_name_lista = list(data_properati['state_name'].unique())
state_name_lista
```

```python
# tipos de propiedades
data_properati['property_type'].value_counts()
```

```python
# genero una mascara para identificar aquellos que tienen el valor de las expensas no nulo
expenses_not_null = data_properati['expenses'].isnull() == 0
# aplico la mascara
data_properati_expenses_not_null = data_properati.loc[expenses_not_null, :]
print(data_properati_expenses_not_null['property_type'].value_counts())
print('\ntotal not null:', data_properati_expenses_not_null['property_type'].count())
```

```python
# quiero ver la distribucion de los not null por state_name
data_properati_expenses_not_null['state_name'].value_counts()
```

```python
# quiero ver el porcentaje de completitud de la variable expensas por state_name
# expensas not null / q total
total_state_name = data_properati['state_name'].value_counts()
porcentaje_expenses_notnull_statename = (((data_properati_expenses_not_null['state_name'].value_counts()) / total_state_name)*100).sort_values(ascending=False)
porcentaje_expenses_notnull_statename
```

```python
# genero una mascara para detectar aquellos datos con expensas no nulas y precio en dolares
# quiero ver si hay correlacion entre el valor de la propiedad y el valor de las expensas 
# me quedo con las propiedades que tienen el precio en dolares
expenses_not_null_usd = (data_properati['expenses'].isnull() == 0) & (data_properati['currency'] == 'USD')
data_properati_expenses_not_null_usd = data_properati.loc[expenses_not_null_usd, :]
print(data_properati_expenses_not_null_usd['state_name'].value_counts())
print('total no nulos con precio en dolares:',data_properati_expenses_not_null_usd['state_name'].count())
```

```python
data_properati_expenses_not_null_usd.loc[:, ['price', 'expenses']].corr(method='pearson')
```

**Conclusión: el valor de la propiedad no correlaciona con el valor de las expensas**
