```python
table_pivote_tipo = df.pivot_table(values = ['price'], columns = 'property_type', aggfunc = completitud)\
    .transpose() \
    .sort_values(by = 'price', ascending = False)

table_pivote_tipo
```

```python
completitud = lambda x: 100 - 100 * x.isnull().sum() / len(x)

table_pivote = df.pivot_table(values = ['price'], columns = 'state_name', aggfunc = completitud)\
    .transpose() \
    .sort_values(by = 'price', ascending = False)

table_pivote['cantidad_de_propiedades'] = df.state_name.value_counts()

table_pivote['promedio_de_precios_m2'] = df.groupby('state_name').agg('mean').price_usd_per_m2

table_pivote['std_de_precios_m2'] = df.groupby('state_name').agg('std').price_usd_per_m2

table_pivote
```

```python
tolerance_table = table_pivote[(table_pivote['price'] >= 74) & (table_pivote['cantidad_de_propiedades'] > 250)]
tolerance_table
```

```python
porcentaje_de_registros = 100 * tolerance_table['cantidad_de_propiedades'].sum() / df.shape[0]
porcentaje_de_registros
```
